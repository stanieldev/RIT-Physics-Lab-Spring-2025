# Scaffolding Notebook for 2D Plotting

This notebook introduces basic 2D plotting techniques using Python's `matplotlib` and `numpy` libraries.
It serves as a resource to understand the core principles of creating plots like line plots, scatter plots,
density plots, and contour plots, and how to effectively use `meshgrid` for 2D grid-based computations.


## Example 1: Basic Line Plot

Line plots are used to display data points connected by a line.


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate data for the line plot
x = np.linspace(0, 10, 100)  # x-axis values from 0 to 10
y = np.sin(x)  # y-axis values as sin(x)

# Create a line plot
plt.plot(x, y, label='$y = sin(x)$')
plt.xlabel('x-axis')  # Label for x-axis
plt.ylabel('y-axis')  # Label for y-axis
plt.title('Line Plot of $y = sin(x)$')  # Title of the plot
plt.grid(True)  # Add gridlines for better readability
plt.legend()  # Add a legend to explain the plot
plt.show()
```

## Example 2: Scatter Plot

Scatter plots are used to show individual data points in a 2D space. They do not connect points with a line.


```python
# Generate structured data points
x = np.array([0, 1, 2, 3, 4, 5])  # x-coordinates
y = np.array([0, 1, 4, 9, 16, 25])  # y-coordinates

# Create a scatter plot
plt.scatter(x, y, color='blue')  # Scatter plot with blue dots
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot of $y = x^2$')
plt.grid(True)
plt.show()
```

## Example 3: Understanding `meshgrid`

The `meshgrid` function generates a grid of coordinates from two 1D arrays of values, $x_i$ and $y_j$, defined as follows:
- $x_i$: The values of the x-coordinates, ranging over $[x_{\text{min}}, x_{\text{max}}]$.
- $y_j$: The values of the y-coordinates, ranging over $[y_{\text{min}}, y_{\text{max}}]$.

For 1D arrays $x_i$ and $y_j$:
$$
x = [x_1, x_2, \dots, x_N] \quad \text{and} \quad y = [y_1, y_2, \dots, y_M]
$$

the `meshgrid` function produces two 2D arrays:
1. **Grid of x-coordinates**:
$$
X = \begin{bmatrix}
x_1 & x_2 & \dots & x_N \\
x_1 & x_2 & \dots & x_N \\
\vdots & \vdots & \ddots & \vdots \\
x_1 & x_2 & \dots & x_N
\end{bmatrix}_{M \times N}
$$
Each row contains the repeated values of $x$.

2. **Grid of y-coordinates**:
$$
Y = \begin{bmatrix}
y_1 & y_1 & \dots & y_1 \\
y_2 & y_2 & \dots & y_2 \\
\vdots & \vdots & \ddots & \vdots \\
y_M & y_M & \dots & y_M
\end{bmatrix}_{M \times N}
$$
Each column contains the repeated values of $y$.

This grid is particularly useful for evaluating a function $f(x, y)$ over a 2D domain, producing a 2D array of function values.


```python
# Define 1D arrays for x and y
x = np.linspace(-1.5, 1.5, 4)  # x-coordinates (e.g., x_1, x_2, ..., x_N)
y = np.linspace(0.5, 2.5, 3)   # y-coordinates (e.g., y_1, y_2, ..., y_M)

# Generate 2D grids for x and y
X, Y = np.meshgrid(x, y)

# Display the grids
print("X (grid of x-coordinates):")
print(X)
print("\nY (grid of y-coordinates):")
print(Y)
```

## Example 4: Understanding How `imshow` Maps a 2D Array

The `imshow` function visualizes a 2D array \( Z_{ij} \) as a grid of colored blocks, where:
- The value \( Z[i, j] \) determines the color of the block at position \((j, i)\) on the plot.
- **Columns (\(j\))** of the array are mapped to the **x-axis** (horizontal axis).
- **Rows (\(i\))** of the array are mapped to the **y-axis** (vertical axis).

### Key Concept: Array-to-Plot Mapping

For a 2D array:
$$
Z = 
\begin{bmatrix}
Z_{11} & Z_{12} & Z_{13} & \dots & Z_{1N} \\
Z_{21} & Z_{22} & Z_{23} & \dots & Z_{2N} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
Z_{M1} & Z_{M2} & Z_{M3} & \dots & Z_{MN}
\end{bmatrix},
$$
each element \( Z[i, j] \):
1. Is displayed as a rectangular block on the plot.
2. The block is positioned at:
   - \( j \): The column index (x-coordinate).
   - \( i \): The row index (y-coordinate).

### Custom Axes with `extent`

By default:
- The **x-axis** runs from 0 to the number of columns (\( N \)).
- The **y-axis** runs from 0 to the number of rows (\( M \)).

Using the `extent` argument, you can map these indices to the intended coordinate ranges:
$$
\text{extent} = [x_{\text{min}}, x_{\text{max}}, y_{\text{min}}, y_{\text{max}}].
$$


### Coarse Mesh Example

This plot visualizes the values of the array $Z$, defined as:
$$
Z[i, j] = x_j^2 + y_i,
$$
where:
- $x_j$ are the values of $x$ along the columns (x-axis).
- $y_i$ are the values of $y$ along the rows (y-axis).

The coarse mesh highlights how the 2D array $Z$ maps to the plot.


```python
# Define the ranges for x and y
x = np.linspace(-1.5, 2.0, 6)  # 7 non-integer values for x
y = np.linspace(0.2, 0.8, 3)  # 3 non-integer values for y

# Create the grid and compute Z = x^2 + y
X, Y = np.meshgrid(x, y)
Z = X**2 + Y

# Plot the array Z using imshow
plt.imshow(Z, cmap='viridis', origin='lower', extent=[x[0], x[-1], y[0], y[-1]])  # Map array to real coordinates
plt.colorbar(label='Value of $x^2 + y$')  # Add a color bar
plt.title('Density Plot of $x^2 + y$')  # Title of the plot
plt.xlabel('x-axis (values of x)')  # Label x-axis
plt.ylabel('y-axis (values of y)')  # Label y-axis
plt.show()

# Print the array Z to show its values
print("The array Z:")
print(Z)
```




### Tighter Mesh Example

This plot visualizes the values of the array $Z$, defined as:
$$
Z[i, j] = x_j^2 + y_i,
$$
where:
- $x_j$ are the values of $x$ along the columns (x-axis).
- $y_i$ are the values of $y$ along the rows (y-axis).

The tighter mesh provides a smoother and more detailed visualization compared to the coarse mesh.


```python
# Define the ranges for x and y (tighter mesh)
x_tight = np.linspace(-1.5, 2.0, 500)  # 50 non-integer values for x
y_tight = np.linspace(0.2, 0.8, 200)  # 20 non-integer values for y

# Create the grid and compute Z = x^2 + y on the tighter mesh
X_tight, Y_tight = np.meshgrid(x_tight, y_tight)
Z_tight = X_tight**2 + Y_tight

# Plot the array Z_tight using imshow
plt.imshow(Z_tight, cmap='viridis', origin='lower', extent=[x_tight[0], x_tight[-1], y_tight[0], y_tight[-1]])  # Map array to real coordinates
plt.colorbar(label='Value of $x^2 + y$')  # Add a color bar
plt.title('Density Plot of $x^2 + y$ (Tighter Mesh)')  # Title of the plot
plt.xlabel('x-axis (values of x)')  # Label x-axis
plt.ylabel('y-axis (values of y)')  # Label y-axis
plt.show()  # Display the plot

# No matrix printout, only visualized plot to highlight the tighter mesh behavior.
```
