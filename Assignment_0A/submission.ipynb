## Problem 1 (25 Points)
A ball is dropped from a tower of height $h$ with initial velocity zero. Write a program that asks the user to enter the height of the tower in meters and then calculates and prints the time the ball takes until it hits the ground, ignoring air resistance. Use your program to calculate the time for a ball dropped from a $100\mathrm{m}$ high tower.

## Theory
Consider a free body under the influence of just gravity, then we can say that:
$$\vec{F}=m\vec{a} \implies m\vec{g}=m\dfrac{d^2\vec{r}}{dt^2} \implies \vec{g}=\dfrac{d^2\vec{r}}{dt^2}$$
Solving the differential equation analytically (with constant acceleration), we get a system of the following form:
$$\vec{x}(t)=\vec{x}_0+\vec{v}_0t+\dfrac{1}{2}\vec{a}t^2$$
Defining the vertical direction as +y, the object is at rest at $t=0$, and that gravity is the only acceleration, we get that:
$$\vec{x}(t)=h-\dfrac{1}{2}\vec{g}t^2$$
To find when the position is $0$, being that of the ground, we get the following relationship:
$$\begin{align}
h-\dfrac{1}{2}gt^2&=0\\
-\dfrac{1}{2}gt^2&=-h\\
\dfrac{1}{2}gt^2&=h\\
t^2&=\dfrac{2h}{g}\\
t&=\sqrt{\dfrac{2h}{g}}
\end{align}$$

```python
HEIGHT = float(input("Enter the height in meters: "))
# HEIGHT = 100      # in meters
CONSTANT_g = 9.81  # in m/s^2
time = (2 * HEIGHT / CONSTANT_g) ** 0.5
print(f"The time it takes for the ball to hit the ground is {time:.3f} seconds (without air resistance).")
```

## Problem 3 (50 Points)
There is a file on myCourses folder “Files and Data” called `stm.txt`, which contains a grid of values from scanning tunneling microscope measurements of the (111) surface of silicon. A scanning tunneling microscope (STM) is a device that measures the shape of a surface at the atomic level by tracking a sharp tip over the surface and measuring quantum tunneling current as a function of position. The end result is a grid of values that represent the height of the surface.

The file stm.txt is organized as follows:
 - Each row corresponds to a unique y-coordinate.
 - Each column corresponds to a unique x-coordinate.
 - The value at a specific row and column represents the height of the surface at that point.

To work with the data:
 - Refer to the notebook Examples-DataDownload.ipynb on myCourses for instructions on how to handle downloading and reading files of this type.
 - Refer to the notebook Examples-Plots.ipynb for a possible approach to visualizing this type of data as a plot. This is not mandatory, and you are encouraged to explore your own methods if you wish.
 - Use numpy.loadtxt to load the data into a 2D array for processing.
 - Generate a visualization of the data that highlights the structure of the silicon surface.

```python
import numpy as np
import matplotlib.pyplot as plt

# Open Data
data = np.loadtxt("stm.txt")

# Data is a 2D array:
# - Each row corresponds to a unqiue y-value
# - Each column corresponds to a unique x-value
# - The value in each cell corresponds to the height of the surface at that (x, y) coordinate
# This is only used to get the endpoints of each data axis
x_data, y_data = np.arange(data.shape[1]), np.arange(data.shape[0])

# Plot the surface using imshow
plt.imshow(data, cmap='viridis', origin='lower', extent=[x_data[0], x_data[-1], y_data[0], y_data[-1]])  # Map array to real coordinates
plt.colorbar(label='Height (nm)')   # Add a color bar
plt.title('Surface of a Material')  # Title of the plot
plt.xlabel('x-axis (nm)')           # Label x-axis
plt.ylabel('y-axis (nm)')           # Label y-axis
plt.show()                          # Display the plot
```
